```r
## create a data object with atomic and recursive features
x <- list(
    character = letters[10],
    numeric = rnorm(10),
    integer = 1:10,
    list_of_char = c(letters, letters[1:9]),
    list_of_dfs = replicate(10, mtcars, simplify = FALSE)
)
```

```r
x
```

```r
## apply is.recursive and is.atomic to each element. notice how is.atomic is the inverse of is.recursive.
sapply(x, is.recursive)
```

```r
sapply(x, is.atomic)
```
