```r
findC <- function(x){
    colNames <- names(x)
    selectCols <- c()
    ind <- 1
    for (i in 1:length(colNames)){
        if (startsWith(colNames[i], "c")){
            selectCols[[i]] <- colNames[i]
            i <- i + 1
        }
    }
    
    return(selectCols)
}

C_Columns <- findC(diamonds)
print(C_Columns)

head(diamonds[, C_Columns])
```

```r
landuse_url <- "https://en.wikipedia.org/wiki/Land_use_statistics_by_country"

landuse <- read_html(landuse_url)
landuse
```

```r
tables <- html_table(landuse)
landuse_df <- tables[[1]]
head(landuse_df)
```

```r
names(landuse_df) <- c("rank", "country", "cultivatedlandkm2", 
                      "cultivatedlandpercent", "arablelandkm2",
                       "arablelandpercent", "permanentcropskm2",
                       "permanentcropspercent", "otherlandskm2",
                       "otherlandspercent", "totalareakm2", "date"                      
                      )

names(landuse_df)
```

```r
head(landuse_df)
```

```r
numericColumns <- 
c("rank", "cultivatedlandkm2", "cultivatedlandpercent",
"arablelandkm2", "arablelandpercent", "permanentcropskm2",
 "permanentcropspercent", "otherlandskm2", "otherlandspercent",
 "totalareakm2", "date")

landuse_df[numericColumns] <- lapply(landuse_df[numericColumns], function(x){
    as.numeric(stringr::str_replace_all(x, "[^[:digit:]\\-\\.]", ""))
})
```

```r
head(landuse_df)
```

```r
interestedCountries <- c("India", "United States", "China", "Russia")

data <- landuse_df %>%
            filter(country %in% interestedCountries) %>%
            select(country, cultivatedlandkm2, arablelandkm2)

data

```

```r
reshapedData <- melt(data, id.vars = "country")
reshapedData
```

```r
ggplot(data = reshapedData, aes(x = country, y = value, fill = variable)) + 
    geom_bar(stat = "identity", position = "dodge")
```
