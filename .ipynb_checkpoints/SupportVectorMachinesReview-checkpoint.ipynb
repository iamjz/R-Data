```r
library(plyr)
library(ggplot2)
library(e1071)
library(caret) 
```

## Quick example

```r
points_data1 <- read.csv("datasets/data1.csv", header = TRUE)
head(points_data1)
```

```r
count(points_data1, "class")
```

```r
points_data1$class <- factor(points_data1$class)
```

```r
plot1 <- ggplot(points_data1, aes(X, Y)) + 
            geom_point(aes(color = factor(class), shape = factor(class)))

plot1
```

```r
svm_model1 <- svm(class ~. ,
                     data = points_data1,
                     kernel = "linear",
                     cost = 10, 
                     scale = FALSE)

summary(svm_model1)
```

```r
plot(svm_model1, points_data1)
```

We trained the SVM with a linear kernel; and it learned a linear decision boundary. The "X" marks in the above plot are the data points that serve as support vectors. There are three of them. Let's compute the confusion matrix, and as expected, we'll get perfect accuracy.

```r
pred <- predict(svm_model1, points_data1[, -3])
```

```r
confusionMatrix(pred, points_data1$class)
```

## A bit more complicated example

```r
points_data <- read.csv("datasets/data3.csv", header = TRUE)
```

```r
head(points_data)
```

```r
count(points_data, "class")
```

```r
points_data$class <- factor(points_data$class)
```

```r
plot1 <- ggplot(points_data, aes(X, Y)) + 
            geom_point(aes(color = factor(class), shape = factor(class)))

plot1
```

```r
svm_model1 <- svm(class ~. ,
                     data = points_data,
                     kernel = "radial",
                     cost = 10, 
                     scale = FALSE)

summary(svm_model1)
```

```r
plot(svm_model1, points_data)
```

```r
pred <- predict(svm_model1, points_data[, -3]) 
```

```r
confusionMatrix(pred, points_data$class)
```

There is a potential problem here. We trained and tested our model with the same exact data set. This can cause memorization; the model does not learn a decision boundary, it memorizes a boundary for this particular data set. We don't know how it will perform on the new, unseen observations. One of the most important aspects of learning algorithms is their ability to generalize; that is, to learn decision boundaries that are generalized enough to do well on unseen data.

### Split test

Split test is simply splitting the data into training and testing sets; 
we can use 65% of the data for training, and the rest for testing 

```r
split <- 0.65
```

```r
trainIndex <- createDataPartition(points_data$class, p = split, list = FALSE)
trainIndex[0:20]
```

```r
train_set <- points_data[trainIndex,]
dim(train_set)
```

```r
test_set <- points_data[-trainIndex,]
dim(test_set)
```

```r
svm_model2 <- svm(class ~. ,
                     data = train_set, 
                     kernel = "radial", 
                     cost = 10, 
                     scale = FALSE)

summary(svm_model2)
```

```r
plot(svm_model2, train_set)
```

```r
predTrain <- predict(svm_model2, train_set[, -3])
predTest <- predict(svm_model2, test_set[, -3])
```

```r
confusionMatrix(predTrain, train_set$class)
```

```r
confusionMatrix(predTest, test_set$class)
```

Usually, the model is expected to do a better job on predictions for the training set; because that's what it has learned. If it has generalized well enough, it should also produce good performance for the testing set. The problem still continues though; how do we know that this particular training set represents the class distribution well enough? We need to repeat this split test a number of times and compute the mean and standard deviation of the accuracy. We can do this with more random splittings, or we can use the cross-validation approach.

### Cross validation

k-fold cross-validation splits the data set into *k* subsets (folds) and then picks one subset for
testing and trains the model with the remaining *k-1* subsets. 
It does this for each fold; so for k=10, it'll end up doing 10 training and testing sessions. 

We can use cross validation to **tune the parameters** of the SVM model to get a better accuracy 
without the danger of memorization. We will use svm tuning functions for that. 

```r
set.seed(129)

# Setup for cross validation:
# sampling="cross" for cross-validation, 
# cross=10 for 10-fold,
# retain the best model and save the performance measures

tctrl <- tune.control(sampling = "cross",
                         cross = 10, 
                         best.model = TRUE,
                         performances = TRUE)
```

```r
# now run the tune function to tune the parameters of the model 
# tune function will try to find the best parameters (gamma and cost), that means the parameters with the smallest error.
# it will try different gamma and cost values given as arguments (e.g. cost=1, cost=10, cost=100, etc.)

tuned_params_cv <- tune(svm, 
                           class ~ ., 
                           data = points_data,
                           kernel = "radial",
                           ranges=list(gamma=10^(-6:-1), cost=10^(0:2)), 
                           tunecontrol = tctrl)

summary(tuned_params_cv)
```

```r
# best one is with an error of 0.044. 
# cost = 100, gamma = 0.1

svm_model_cv <- svm(class ~ ., 
                       data = points_data,
                       kernel = "radial",
                       cost = 100, 
                       gamma = 0.1,
                       scale = FALSE)

summary(svm_model_cv)
```

```r
plot(svm_model_cv, points_data)
```

```r
pred <- predict(svm_model_cv, points_data[, -3])
```

```r
confusionMatrix(pred, points_data$class)
```

### But we should really be using the testing and training sets here. 

```r


svm_model_cv2 <- svm(class ~., 
                        data = train_set, 
                        kernel = "radial", 
                        cost = 100,
                        gamma = 0.1,
                        scale = FALSE)

summary(svm_model_cv2)
```

```r
plot(svm_model_cv2, train_set)
```

```r
predTrain <- predict(svm_model_cv2, train_set[, -3])
predTest <- predict(svm_model_cv2, test_set[, -3])
```

```r
confusionMatrix(predTrain, train_set$class)
```

```r
confusionMatrix(predTest, test_set$class)
```

## Last example

```r
points_data <- read.csv("datasets/data4.csv", header = TRUE)
```

```r
head(points_data)
```

```r
points_data$class <- factor(points_data$class)
```

```r
plot1 <- ggplot(points_data, aes(X, Y)) + 
            geom_point(aes(color = factor(class), shape = factor(class)))

plot1
```

```r
tctrl <- tune.control(sampling = "cross", 
                         cross = 10, 
                         best.model = TRUE,
                         performances = TRUE)
```

```r
tuned_params_cv <- tune(svm, class ~. ,
                           data = points_data,
                           kernel = "radial",
                           ranges = list(gamma = 10^(-6:-1), cost=10^(0:2)), 
                           tunecontrol = tctrl)

summary(tuned_params_cv)
```

```r
split <- 0.65

trainIndex <- createDataPartition(points_data$class, p = split, list = FALSE)
```

```r
train_set <- points_data[trainIndex, ]
dim(train_set)
```

```r
test_set <- points_data[-trainIndex, ]
dim(test_set)
```

```r
svm_model <- svm(class ~.,
                    data = train_set,
                    kernel = "radial",
                    gamme = 0.1,
                    cost = 100, 
                    scale = FALSE)

summary(svm_model)
```

```r
plot(svm_model, train_set)
```

```r
predTrain <- predict(svm_model, train_set[, -3])
predTest <- predict(svm_model, test_set[, -3])
```

```r
confusionMatrix(predTrain, train_set$class)
```

```r
confusionMatrix(predTest, test_set$class)
```
