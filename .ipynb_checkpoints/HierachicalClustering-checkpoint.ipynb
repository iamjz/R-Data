```r
library(ggplot2)
library(plyr)
```

# Example 1

```r
points_data1 <- read.csv("datasets/data1.csv")
```

```r
head(points_data1)
```

```r
plot1 <- ggplot(points_data1, aes(X, Y)) + 
            geom_point(colour="black")

plot1
```

```r
set.seed(129)

hc_clust11 <- hclust(dist(points_data1[, 1:2]), 
                         method = "complete")

plot(hc_clust11)
```

In a dendrogram, the height gives a good idea about how "far" the clusters are in terms of dissimilarity. 
Above, it seems like there are two clusters in this data set based on the heights of the tree branches. 
Let's cut it at two clusters and visualize them.

```r
cutree(hc_clust11, k = 2)
```

```r
plot(hc_clust11)

depth.cutoff <- 8
abline(h = depth.cutoff, col="red", lty = 2)
```

```r
cut2 <- cutree(hc_clust11, 2)
cut2
```

```r
plot1 <- ggplot(points_data1, aes(X, Y)) + 
            geom_point(aes(colour = factor(cut2))) + 
            theme(legend.position = "top")

plot1
```

## Example 1 Data - Using factoextra library

```r
library(factoextra)

# run k-means on this data and visualize 

km <- eclust(points_data1[, 1:2], 
                 "kmeans", 
                 k = 2, 
                 nstart = 20, 
                 graph = FALSE)


km
```

```r
fviz_cluster(km, geom="point", frame=FALSE)
```

```r
hc <- eclust(points_data1[, 1:2], "hclust", k=2, method="complete", graph=FALSE) 
hc
```

```r
fviz_cluster(hc, geom = "point", frame = TRUE)
```

```r
fviz_dend(hc, rect = TRUE, show_labels = FALSE) 
```

```r
table(cut2, points_data1$class)
```

```r
# another way 
table(hc$cluster, points_data1$class)
```

```r
points_data2 <- read.csv("datasets/data2.csv", 
                             header = TRUE)

# Visualize the data
plot2 <- ggplot(points_data2, aes(X, Y)) + 
            geom_point(colour = "black")

plot2
```

```r
head(points_data2)
```

```r
count(points_data2, "class")
```

```r
points_data2$class[points_data2$class == -1] <- 2
points_data2$class[points_data2$class == 1] <- 1
```

```r
count(points_data2, "class")
```

```r
set.seed(129)

# run hclust on this data and visualize 
hc2 <- eclust(points_data2[, 1:2], 
                  "hclust", 
                  k = 2, 
                  method = "complete", 
                  graph = FALSE)

hc2
```

```r
fviz_cluster(hc2, geom = "point", frame = TRUE)
```

```r
# Dendrogram
fviz_dend(hc2, rect = TRUE, show_labels = FALSE)
```

```r
cut2 <- cutree(hc2, 2)
```

```r
cut2
```

```r
table(cut2, points_data2$class)
```
