### Find optimal eps value and nunber of nearest neighbours to be considered for forming clusters. 

```r
library(dbscan)
library(fpc)
library(factoextra)
```

```r
kNNdistplot(points_data2[,1:2], k =  4)
abline(h = 0.4, lty = 2,col='red')
```

```r
set.seed(129)

input <- points_data2[, 1:2]

db <- fpc::dbscan(input, 
                  eps = 0.4, 
                  MinPts = 4)

# Plot DBSCAN results
plot(db, 
         input, 
         main = "DBSCAN", 
         frame = FALSE)
```

DBSCAN did found two major clusters but there are many small clusters which are invalid. Many outliers are detected. Plot the points using fviz_cluster()

```r
fviz_cluster(db, input, 
                stand = FALSE, 
                frame = FALSE, 
                geom = "point")
```

```r
db <- fpc::dbscan(input, 
                      eps = 0.5, 
                      MinPts = 4)


# Plot DBSCAN results
plot(db, input, 
             main = "DBSCAN", 
             frame = FALSE)
```

# Another example

```r
# read data from the file
points_data3 <- read.csv("datasets/data3.csv",header=TRUE)

str(points_data3)

# Visualize the data
library(ggplot2)
plot1 <- ggplot(points_data3, aes(X, Y)) + geom_point(colour="black")
plot1
```

```r
input <- points_data3[, 1:2]
```

```r
db <- fpc::dbscan(input, 
                      eps = 0.25, 
                      MinPts = 3)

# Plot DBSCAN results
plot(db, input, 
         main = "DBSCAN", 
         frame = FALSE)
```

```r
fviz_cluster(db, input, 
                 stand = FALSE, 
                 frame = FALSE, 
                 geom = "point")
```

```r
db <- fpc::dbscan(input, 
                      eps = 0.2, 
                      MinPts = 5)

# Plot DBSCAN results
plot(db, input, main = "DBSCAN", frame = FALSE)
```

dbscan was able to find 6 clusters with a smaller eps value. The clusters are overlapping, have different densities, and have outliers; dbscan can easily merge clusters for a slightly larger eps, or produce many outliers for a slightly smaller eps for this data set.
