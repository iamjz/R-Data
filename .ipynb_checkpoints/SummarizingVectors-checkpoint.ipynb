We have to identify independent variables that are related to our response variable, "price". To do this, we will look for bivariate relationships. 

We know, culturally, that bedrooms will be a major decider in the price of a house. So, let's look at this and test our expectation with a scatterplot of the price and bedrooms. We will add a regression line to our scatterplot as well so we can estimate the correlation coefficient between the variables. 

```r
ggplot(housing_prices, aes(x = bedrooms, y = price / 1000)) + 
    xlab("Bedrooms") + 
    ylab("Price ($K)") + 
    geom_point() +
    geom_smooth(method = lm) 
    # add the smooth geometry element with a linearly model
```

This is our basic plot. Somethign that often helps in plots is to bring in another feature of the data into the plot via the use of colors. Let's redo this...

```r
ggplot(housing_prices, aes(x = bedrooms, y = price/1000)) +
    xlab("Bedrooms") + 
    ylab("Price ($K)") + 
    geom_point(aes(color = view)) + 
    geom_smooth(method = lm)
```

```r
ggplot(housing_prices, aes(x = bedrooms, y = price/1000)) + 
    xlab("Bedrooms") + 
    ylab("Price ($K)") +
    geom_point(aes(color = floors)) + 
    geom_smooth(method = lm)
```

33 bedrooms actually does look like an outlier because its price is similar to what a 4 bedroom house will cost. Look at the particular record in the dataset.

```r
housing_prices[housing_prices$bedrooms == 33, ]
```

From a glance, it looks like an input error. We're going to remove it from our dataset. Fuck outliers!

```r
housing_prices <- housing_prices[!housing_prices$bedrooms %in% c(33),]
```

```r
ggplot(housing_prices, aes(x = bedrooms, y = (price/1000))) + 
    xlab("Bedrooms") + 
    ylab("Price ($K)") + 
    geom_point(aes(color = view))  +
    geom_smooth(method = lm)
```

## Linear regression model

we'll use the linear model solver lm(). 

The LM function computes a linear regression
* LHS is the dependent varible
* RHS are is the independent variables (predictors

```r
fit <- lm(price ~ sqft_living + grade, data = hp)
```

```r
summary(fit)
```

#### NOTE: Multiple R-squared:  0.5345,	Adjusted R-squared:  0.5345 

What if we add another predictor? sqft_above?

```r
fit3 <- lm(price ~ sqft_living + grade + sqft_above, data = hp)
summary(fit3)
```

```r
fit4 <- lm(price ~ sqft_living + grade + sqft_above + bathrooms, data = hp)
summary(fit4)
```

We can see that as we add predictors with lower correlations to price, we get diminishing returns on the R squared measure of fitness. 

```r
fit
```

```r
library(ggplot2)

########################
#  adapted from: 
# https://susanejohnston.wordpress.com/2012/08/09/a-quick-and-easy-function-to-plot-lm-results-in-r/
########################

    # Note that fit is an object with various things, such as a model.
ggplot(hp, 
       aes_string(
                x = (184.4*hp$sqft_living + 98559.0*hp$grade - 598157.0), 
                y = names(fit$model)[1] # price
       ) # end of aes_string
  ) + 
  geom_point() +
  stat_smooth(method = "lm", col = "red") +
  labs(title = paste("Adj R2 = ",signif(summary(fit)$adj.r.squared, 5),
                     "Intercept =",signif(fit$coef[[1]],5 ),
                     " Slope =",signif(fit$coef[[2]], 5),
                     " P =",signif(summary(fit)$coef[2,4], 5)
                ) # end of title string concatenations
       , x = "sqft_living + grade"
      ) 
```
